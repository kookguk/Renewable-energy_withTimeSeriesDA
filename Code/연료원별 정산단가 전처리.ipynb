## 1. 라이브러리 호출 및 데이터 불러오기

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

df=pd.read_csv('연료원별 정산단가.csv')
df
```

## 2. 22~23년도 데이터만 남긴 후 날짜 형식으로 변환

```python
df = df.dropna(subset=['기간'])

df = df[df['기간'].str.endswith(('22', '23'))]

# '월.연도' 형식을 'yyyy-mm' 형식으로 변환하는 함수
def convert_to_date_format(period):
    month_map = {
        'Jan': '01', 'Feb': '02', 'Mar': '03', 'Apr': '04',
        'May': '05', 'Jun': '06', 'Jul': '07', 'Aug': '08',
        'Sep': '09', 'Oct': '10', 'Nov': '11', 'Dec': '12'
    }
    month_str, year_str = period.split('.')
    year = '20' + year_str
    month = month_map[month_str]
    return f'{year}-{month}'

df['기간'] = df['기간'].apply(convert_to_date_format)

df
```

## 3. 월별->일별로 확장 및 월별로 같은 값 적용

```python
# '기간' 칼럼을 일별 데이터로 확장하는 함수
def expand_month_to_days(row):
    period = row['기간']
    start_date = datetime.strptime(period, "%Y-%m")
    end_date = (start_date + pd.DateOffset(months=1)) - timedelta(days=1)
    
    all_days = pd.date_range(start=start_date, end=end_date, freq='D')
    
    expanded_df = pd.DataFrame({
        '기간': all_days
    })
    
    for col in df.columns:
        if col != '기간':
            expanded_df[col] = row[col]
    
    return expanded_df

expanded_dfs = [expand_month_to_days(row) for _, row in df.iterrows()]

result_df = pd.concat(expanded_dfs, ignore_index=True)

result_df = result_df.sort_values(by='기간').reset_index(drop=True)

result_df
```
