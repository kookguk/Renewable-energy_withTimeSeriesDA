## 1. 워드클라우드 단어 count 파일, 감성분석 최고 기여 단어 파일 병합, 약간의 전처리

```python
import pandas as pd

df1 = pd.read_excel('제주도 에너지 뉴스제목 워드클라우드.xlsx')
df2 = pd.read_excel('제주도 에너지 감성분석 최고 기여 단어.xlsx')

# 'text' 열과 'top_words' 열을 기준으로 병합
merged_df = pd.merge(df1, df2, left_on='text', right_on='top_word', how='inner')

# 중복된 행 제거
merged_df.drop_duplicates(inplace=True)

# 'text' 칼럼 삭제
merged_df.drop(columns=['text'], inplace=True)

# 칼럼 순서 조정
merged_df = merged_df.reindex(columns=['top_word', 'count', 'label', 'confidence'])

# 결과를 엑셀 파일로 저장
merged_df.to_excel('제주도 에너지 워드클라우드, 감성분석 병합.xlsx', index=False)
```
